```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data
df = pd.read_csv("income.csv")

# Features for clustering
X = df[['Age','Income($)']]

# Create model
kmc = KMeans(n_clusters=3, random_state=0)

# Fit model
df['Cluster'] = kmc.fit_predict(X)

# Plot clusters
plt.scatter(df['Age'], df['Income($)'], c=df['Cluster'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("income.csv")

# Select features
X = df[['Age', 'Income($)']]

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
km = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = km.fit_predict(X_scaled)

# Plot clusters
plt.scatter(df['Age'], df['Income($)'], c=df['Cluster'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```
