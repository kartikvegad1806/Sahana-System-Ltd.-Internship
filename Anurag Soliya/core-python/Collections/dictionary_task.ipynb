```python
my_int_list = []
my_str_list = []


while (1):
    inp = (input("enter anything: "))
    if inp.isnumeric():
        if int(inp) == 0:
            break
        my_int_list.append(int(inp))
    elif inp.isalpha():
        my_str_list.append(inp)

print(my_int_list)
print(my_str_list)


print("minimum in int list is: {}".format(min(my_int_list)))
print("maximum in int list is: {}".format(max(my_int_list)))

my_str_list.reverse()

print(my_str_list)
```

```python
my_dict = {}

while (1):
    inp = input("enter student number: ")
    if int(inp) == 0:
        break
    my_dict[inp] = {
        "name" : input("enter name: "),
        "age" : int(input("enter age: ")),
        "marks" : int(input("enter marks: ")),
        "grade" : None
    }


print(my_dict)
for my_dict_id in my_dict:
    for key, value in my_dict[my_dict_id].items():
        if key == 'marks':
            marks = value
            if  marks <= 100 and marks >= 90:
                my_dict[my_dict_id]['grade'] = 'A'
            elif marks < 90 and marks >= 80:
                my_dict[my_dict_id]['grade'] = 'B'
            elif marks < 80 and marks >= 60:
                my_dict[my_dict_id]['grade'] = 'C'
            elif marks < 60 and marks >= 40:
                my_dict[my_dict_id]['grade'] = 'D'
            else:
                my_dict[my_dict_id]['grade'] = 'Fail'


print("\n----------- my_dict Records -----------")
for my_dict_id, details in my_dict.items():
    print(f"\n{my_dict_id}:")
    for key, value in details.items():
        print(f"  {key}: {value}")
```
